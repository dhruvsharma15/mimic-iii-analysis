## Age Stratification

```python
# Get ages of all patients from patients, admissions table
query = \
"""
SELECT
  p.subject_id, p.dob, p.gender
  , MIN (a.admittime) AS first_admittime
  , MIN( ROUND( (cast(admittime as date) - cast(dob as date)) / 365.242,2) )
      AS first_admit_age
FROM patients p
INNER JOIN admissions a
ON p.subject_id = a.subject_id
GROUP BY p.subject_id, p.dob, p.gender
ORDER BY p.subject_id
"""

data = pd.read_sql_query(query,con)
```

### Grouping on age groups of 10 years

```python
query = \
"""
WITH first_admission_time AS
(
  SELECT
      p.subject_id, p.dob, p.gender
      , MIN (a.admittime) AS first_admittime
      , MIN( ROUND( (cast(admittime as date) - cast(dob as date)) / 365.242,2) )
          AS first_admit_age
  FROM patients p
  INNER JOIN admissions a
  ON p.subject_id = a.subject_id
  GROUP BY p.subject_id, p.dob, p.gender
  ORDER BY p.subject_id
)
, age as
(
  SELECT
      subject_id, dob, gender
      , first_admittime, first_admit_age
      , CASE
          -- all ages > 89 in the database were replaced with 300
          -- we check using > 100 as a conservative threshold to ensure we capture all these patients
          WHEN first_admit_age > 100
              then '>89'
          WHEN first_admit_age >= 80
              THEN '80-89'
          WHEN first_admit_age >= 70
              THEN '70-79'
          WHEN first_admit_age >= 60
              THEN '60-69'
          WHEN first_admit_age >= 50
              THEN '50-59'
          WHEN first_admit_age >= 40
              THEN '40-49'
          WHEN first_admit_age >= 30
              THEN '30-39'
          WHEN first_admit_age >= 20
              THEN '20-29'
          WHEN first_admit_age >= 14
              THEN '14-19'
          WHEN first_admit_age <= 1
              THEN '<=1'
          ELSE 'middle'
          END AS age_group
  FROM first_admission_time
)
select age_group
  , count(subject_id) as NumberOfPatients
from age
group by age_group
order by age_group
"""

data = pd.read_sql_query(query,con)
```

### Grouping on 'neonate', 'adults', and '>89'

```python
query = \
"""
WITH first_admission_time AS
(
  SELECT
      p.subject_id, p.dob, p.gender
      , MIN (a.admittime) AS first_admittime
      , MIN( ROUND( (cast(admittime as date) - cast(dob as date)) / 365.242,2) )
          AS first_admit_age
  FROM patients p
  INNER JOIN admissions a
  ON p.subject_id = a.subject_id
  GROUP BY p.subject_id, p.dob, p.gender
  ORDER BY p.subject_id
)
, age as
(
  SELECT
      subject_id, dob, gender
      , first_admittime, first_admit_age
      , CASE
          -- all ages > 89 in the database were replaced with 300
          -- we check using > 100 as a conservative threshold to ensure we capture all these patients
          WHEN first_admit_age > 100
              then '>89'
          WHEN first_admit_age >= 14
              THEN 'adult(>=14)'
          WHEN first_admit_age <= 1
              THEN 'neonate(<=1)'
          ELSE 'middle'
          END AS age_group
  FROM first_admission_time
)
select age_group
  , count(subject_id) as NumberOfPatients
from age
group by age_group
order by age_group

"""

data = pd.read_sql_query(query,con)
data
```

## Categorization for diagnosis based on ICD9 codes hierarchy from DIAGNOSES_ICD

```python
query = \
'''
select
    case
        when substring(icd9_code,1,1) = 'E'
            then '(E and V)external causes of injury and supplemental classification'
        when substring(icd9_code,1,1) = 'V'
            then '(E and V)external causes of injury and supplemental classification'
        when cast(substring(icd9_code,1,3) as INT) >= 800
            then '(800-999)injury and poisoning'
        when cast(substring(icd9_code,1,3) as INT) >= 780
            then '(780-799)symptoms, signs, and ill-defined conditions'
        when cast(substring(icd9_code,1,3) as INT) >= 760
            then '(760-779)certain conditions originating in the perinatal period'
        when cast(substring(icd9_code,1,3) as INT) >= 740
            then '(740-759)congenital anomalies'
        when cast(substring(icd9_code,1,3) as INT) >= 710
            then '(710-739)diseases of the musculoskeletal system and connective tissue'
        when cast(substring(icd9_code,1,3) as INT) >= 680
            then '(680-709)diseases of the skin and subcutaneous tissue'
        when cast(substring(icd9_code,1,3) as INT) >= 630
            then '(630-679)complications of pregnancy, childbirth, and the puerperium'
        when cast(substring(icd9_code,1,3) as INT) >= 580
            then '(580-629)diseases of the genitourinary system'
        when cast(substring(icd9_code,1,3) as INT) >= 520
            then '(520-579)diseases of the digestive system'
        when cast(substring(icd9_code,1,3) as INT) >= 460
            then '(460-519)diseases of the respiratory system'
        when cast(substring(icd9_code,1,3) as INT) >= 390
            then '(390-459)diseases of the circulatory system'
        when cast(substring(icd9_code,1,3) as INT) >= 320
            then '(320-389)diseases of the nervous system and sense organs'
        when cast(substring(icd9_code,1,3) as INT) >= 290
            then '(290-319)mental disorders'
        when cast(substring(icd9_code,1,3) as INT) >= 280
            then '(280-289)diseases of the blood and blood-forming organs'
        when cast(substring(icd9_code,1,3) as INT) >= 240
            then '(240-279)endocrine, nutritional and metabolic diseases, and immunity disorders'
        when cast(substring(icd9_code,1,3) as INT) >= 140
            then '(140-239)neoplasms'
        when cast(substring(icd9_code,1,3) as INT) >= 1
            then '(001-139)infectious and parasitic diseases'
        else 'unknown'
    end as code_cat, count(*)
from diagnoses_icd
group by code_cat
order by code_cat;
'''
data = pd.read_sql_query(query,con)
data
```

## Categorization for procedures based on ICD9 codes hierarchy from PROCEDURES_ICD

```python
query = \
'''
select
    case
        when cast(substring(icd9_code,1,2) as INT) >= 87
            then '(87+)Miscellaneous diagnostic and therapeutic procedures'
        when cast(substring(icd9_code,1,2) as INT) >= 85
            then '(85-86)Operations on the integumentary system'
        when cast(substring(icd9_code,1,2) as INT) >= 76
            then '(76-84)Operations on the musculoskeletal system'
        when cast(substring(icd9_code,1,2) as INT) >= 72
            then '(72-75)Obstetrical procedures'
        when cast(substring(icd9_code,1,2) as INT) >= 65
            then '(65-71)Operations on the female genital organs'
        when cast(substring(icd9_code,1,2) as INT) >= 60
            then '(60-64)Operations on the male genital organs'
        when cast(substring(icd9_code,1,2) as INT) >= 55
            then '(55-59)Operations on the urinary system'
        when cast(substring(icd9_code,1,2) as INT) >= 42
            then '(42-54)Operations on the digestive system'
        when cast(substring(icd9_code,1,2) as INT) >= 40
            then '(40-41)Operations on the hemic and lymphatic system'
        when cast(substring(icd9_code,1,2) as INT) >= 35
            then '(35-39)Operations on the cardiovascular system'
        when cast(substring(icd9_code,1,2) as INT) >= 30
            then '(30-34)Operations on the respiratory system'
        when cast(substring(icd9_code,1,2) as INT) >= 21
            then '(21-29)Operations on the nose, mouth and pharynx'
        when cast(substring(icd9_code,1,2) as INT) >= 18
            then '(18-20)Operations on the ear'
        when cast(substring(icd9_code,1,2) as INT) >= 8
            then '(8-16)Operations on the eye'
        when cast(substring(icd9_code,1,2) as INT) >= 6
            then '(6-7)Operations on the endocrine system'
        when cast(substring(icd9_code,1,2) as INT) >= 1
            then '(1-5)Operations on the nervous system'
        when cast(substring(icd9_code,1,2) as INT) >= 0
            then '(00)Procedures and interventions, not elsewhere classified'
        else 'unknown'
    end as code_cat, count(*)
from procedures_icd
group by code_cat
order by code_cat;
'''
data = pd.read_sql_query(query,con)
data
```

## Categorozation into the 6 required ICD9 codes groups

```python
query = \
"""
WITH icd_cat as
(
  SELECT
      subject_id, hadm_id
      , CASE
          WHEN icd9_code like '890%' or icd9_code like '896%' or icd9_code like '897%'
              then 'Hemorrhage Control and Tourniquet'
          WHEN icd9_code like '9950%' or icd9_code like '9956%'
              THEN 'Acute Airway Obstruction and Treatment'
          WHEN icd9_code like '860%'
              THEN 'Tension Pneumothorax'
          WHEN icd9_code like '942%' or icd9_code like '943%' or icd9_code like '945%' or icd9_code like '946%' or icd9_code like '948%'
              THEN 'Burn Management'
          WHEN icd9_code like '927%' or icd9_code like '928%'
              THEN 'Crushing Injury'
          WHEN icd9_code like '9916%' or icd9_code like '9920%' or icd9_code like '9923%'
              THEN 'Environmental'
          ELSE 'other categories'
          END AS cat
  FROM diagnoses_icd
)
select cat as ICD_Category
  , count(subject_id) as NumberOfCases
from icd_cat
group by cat
order by cat
"""

data = pd.read_sql_query(query,con)
data
```
